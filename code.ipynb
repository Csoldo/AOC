```python
with open('in.txt') as f:
    lines = f.read() 
lines = lines.split('\n')
formatted_lines = []
for line in lines: 
    formatted_lines.append(line.split(":")[1])

# Solution 1
sum = 0
for card in formatted_lines:
    winning_numbers = card.split("|")[0].split()
    numbers = card.split("|")[1].split()
    wins = 0
    for num in numbers:
        if winning_numbers.__contains__(num):
            wins += 1

    if wins > 0:
        sum += 2**(wins-1)
print(sum)

# Solution 2
number_of_cards = len(formatted_lines)
copies = []
for i in range(number_of_cards):
    copies.append(0)

for i, card in enumerate(formatted_lines):
    winning_numbers = card.split("|")[0].split()
    numbers = card.split("|")[1].split()
    wins = 0
    for num in numbers:
        if winning_numbers.__contains__(num):
            wins += 1
    if wins > 0:
        for j in range(i+1, i+1+wins):
            if j < number_of_cards:
                copies[j] += 1 + copies[i]
print(copies)
sum = number_of_cards
for val in copies:
    sum += val
print(sum)
```

```python
with open("in.txt") as f:
    lines = f.read() 
lines = lines.split('\n')

hands = []
bids = []
for line in lines:
    hands.append(line.split()[0])
    bids.append(int(line.split()[1]))

point_maps = {
    "T": 10,
    "J": 1,
    "Q": 11,
    "K": 12,
    "A": 13
}

def get_hand_points(hand: str):
    hist = [0] * 13
    jokers = 0
    for card in hand:
        if card.isdigit():
            hist[int(card)-1] += 1
        elif card == "J":
            jokers += 1
        else:
            hist[point_maps[card]-1] += 1
    
    hist.sort(reverse=True)
    hist[0] += jokers
    if hist[0] == 5:
        return 6
    elif hist[0] == 4:
        return 5
    elif hist[0] == 3 and hist[1] == 2:
        return 4
    elif hist[0] == 3:
        return 3
    elif hist[0] == 2 and hist[1] == 2:
        return 2
    elif hist[0] == 2:
        return 1
    return 0

def compare_cards(card_a: str, card_b: str):
    card_a_points = get_hand_points(card_a)
    card_b_points = get_hand_points(card_b)
    if card_a_points == card_b_points:
        for i in range(5):
                value_a = int(card_a[i]) if card_a[i].isdigit() else point_maps[card_a[i]]
                value_b = int(card_b[i]) if card_b[i].isdigit() else point_maps[card_b[i]]
                if value_a != value_b:
                    return value_a > value_b
                
    else:
        return card_a_points > card_b_points
    
ranks = []
size = len(hands)
for i, hand in enumerate(hands):
    hands_beaten = 0
    for j in range(size):
        if compare_cards(hand, hands[j]):
            hands_beaten += 1
    ranks.append(hands_beaten+1)

print(ranks)
sum = 0
for i in range(size):
    sum += ranks[i] * bids[i]

print(sum)
```
