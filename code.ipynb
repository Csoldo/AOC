```python
with open('input1.txt') as f:
    lines = f.readlines()

sum = 0
for line in lines:
    blues = 0
    reds = 0
    greens = 0
    words = line.split(" ")
    valid = True
    for i, value in enumerate(words):
        if value.__contains__("blue"):
            blues += int(words[i-1])
        if value.__contains__("red"):
            reds += int(words[i-1])
        if value.__contains__("green"):
            greens += int(words[i-1])
  
        if reds > 12 or greens > 13 or blues > 14:
            valid = False
            break
        if value.__contains__(";"):
            blues = 0
            reds = 0
            greens = 0

    if valid:
        sum += int(words[1].split(":")[0])
print(sum)


#----- PART 2

sum = 0
for line in lines:
    blues = 0
    reds = 0
    greens = 0
    words = line.split(" ")
    max_blue = 0
    max_red = 0
    max_green = 0
    print(words)
    for i, value in enumerate(words):
        if value.__contains__("blue"):
            blues += int(words[i-1])
        if value.__contains__("red"):
            reds += int(words[i-1])
        if value.__contains__("green"):
            greens += int(words[i-1])
  
        if value.__contains__(";") or value.__contains__("\n"):
            max_blue = blues if max_blue < blues else max_blue 
            max_red = reds if max_red < reds else max_red 
            max_green = greens if max_green < greens else max_green
            # print(f"val: {value}, blues: {max_blue}, reds: {max_red}, greens: {max_green}")
            blues = 0
            reds = 0
            greens = 0

    sum += max_green * max_blue * max_red
    print(f"sum: {sum}, blues: {max_blue}, reds: {max_red}, greens: {max_green}")
print(sum)
```

```python
from collections import defaultdict
with open('in.txt') as f:
    lines = f.read() 
lines = lines.split('\n')
G = [[c for c in line] for line in lines]
R = len(G)
C = len(G[0])

p1 = 0
nums = defaultdict(list)
for r in range(len(G)):
  gears = set()
  n = 0
  has_part = False
  for c in range(len(G[r])+1):
    if c<C and G[r][c].isdigit():
      n = n*10+int(G[r][c])
      for rr in [-1,0,1]:
        for cc in [-1,0,1]:
          if 0<=r+rr<R and 0<=c+cc<C:
            ch = G[r+rr][c+cc]
            if not ch.isdigit() and ch != '.':
              has_part = True
            if ch=='*':
              gears.add((r+rr, c+cc))
    elif n>0:
      for gear in gears:
        nums[gear].append(n)
      if has_part:
        p1 += n
      n = 0
      has_part = False
      gears = set()

print(p1)
p2 = 0
for k,v in nums.items():
  if len(v)==2:
    p2 += v[0]*v[1]
print(p2)
```

## DAY 4

```python
with open('in.txt') as f:
    lines = f.read() 
lines = lines.split('\n')
formatted_lines = []
for line in lines: 
    formatted_lines.append(line.split(":")[1])

# Solution 1
sum = 0
for card in formatted_lines:
    winning_numbers = card.split("|")[0].split()
    numbers = card.split("|")[1].split()
    wins = 0
    for num in numbers:
        if winning_numbers.__contains__(num):
            wins += 1

    if wins > 0:
        sum += 2**(wins-1)
print(sum)

# Solution 2
number_of_cards = len(formatted_lines)
copies = []
for i in range(number_of_cards):
    copies.append(0)

for i, card in enumerate(formatted_lines):
    winning_numbers = card.split("|")[0].split()
    numbers = card.split("|")[1].split()
    wins = 0
    for num in numbers:
        if winning_numbers.__contains__(num):
            wins += 1
    if wins > 0:
        for j in range(i+1, i+1+wins):
            if j < number_of_cards:
                copies[j] += 1 + copies[i]
print(copies)
sum = number_of_cards
for val in copies:
    sum += val
print(sum)
```

## DAY 5

```python
with open('in.txt') as f:
    lines = f.read() 
lines = lines.split('\n')

seeds = [int(x) for x in lines[0].split(":")[1].split()]
values_dict = {}
size_limit = len(lines)

i = 2
while i < size_limit:
    map_key = lines[i].split()[0]
    i += 1
    values_dict[map_key] = []
    while lines[i] != "":
        values_dict[map_key].append([int(x) for x in lines[i].split()])
        i += 1
    i += 1


# print(values_dict)
locations = []
for i, seed_index in enumerate(seeds):
    if i % 2 == 1: 
        continue
    for seed in range(seed_index, seed_index + seeds[i+1]):
        current_map = seed
        for map_value in values_dict:
            for map in values_dict[map_value]:
                dest = map[0]
                source = map[1]
                shift = map[2]
                if current_map >= source and current_map < source + shift:
                    diff = dest - source
                    current_map += diff
                    break
            # print(f"seed: {seed}, map: {current_map}")

        locations.append(current_map)

print(locations)
print(min(locations))
```

```python
with open('in.txt') as f:
    lines = f.read() 
lines = lines.split('\n')

# PART 1
times = [int(x) for x in lines[0].split(":")[1].split()]
records = [int(x) for x in lines[1].split(":")[1].split()]

print(times)

product = 1
for i, time in enumerate(times):
    record = records[i]
    cnt = 0
    for k in range(1, time-1):
        dist = k * (time - k)
        if dist > record:
            cnt += 1
    if cnt > 0:
        product *= cnt

print(product)

# PART 2
times = lines[0].split(":")[1].split()
records = lines[1].split(":")[1].split()
time = int("".join(times))
record = int("".join(records))
product = 1
cnt = 0
for k in range(1, time-1):
    dist = k * (time - k)
    if dist > record:
        cnt += 1
if cnt > 0:
    product *= cnt

print(product)

#440000
#26187338
```
